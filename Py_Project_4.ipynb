```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x).
def f(x):
    return np.sqrt(x**2 + 5)

# Plot the function f(x), np.linspace might be helpful.
x = np.linspace(-5, 5, 400)
y = f(x)

# Find the minimum
min_index = np.argmin(y)
x_min = x[min_index]
y_min = y[min_index]

plt.figure()
plt.plot(x, y, label = '$f(x) = \sqrt{x^2+5}$')
plt.plot(x_min, y_min, 'ro', markersize=8, label='Minimum')
plt.title("$f(x) = \sqrt{x^2+5}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the derivative of the function f(x), for gradient descent
def df(x):
    return x / np.sqrt(x**2 + 5)

# Write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.
def gradient_descent(x_0, step_size, num_iterations):
    x_k = np.zeros(num_iterations + 1)
    x_k[0] = x_0
    
    for k in range(num_iterations):
        x_k[k+1] = x_k[k] - step_size * df(x_k[k])
        print(f"x_{k+1} = {x_k[k+1]}")

    print("\nx_T =", x_k[-1])
    return x_k
# Set up/assign initial values
num_iterations = 50
step_size = 1
x_0 = 2/5 + 1 # ID ends with 2

# Run gradient descent by calling the function and report x_T by printing out the value
gd = gradient_descent(x_0, step_size, num_iterations)

# Plot the f(x_k) values vs. iteration number
f_vals = f(gd)
plt.figure()
plt.plot(f_vals)
plt.title("$f(x_k)$ vs. Iterations")
plt.xlabel("Iteration k")
plt.ylabel("$f(x_k)$")
plt.grid(True)
plt.show()
```

```python
# create a list containing step sizes
step_sizes = [5, 3, 1, 0.5]

# initialize a list to store the results from different step sizes
results = {}

# loop through the step sizes and perform GD and record results
for step_size in step_sizes:
    gd = gradient_descent(x_0, step_size, num_iterations)
    fs = f(gd)
    diff = fs[:-1] - fs[1:]
    results[step_size] = diff

# plot multiple curves in a single graph
for step_size, diff in results.items():
    plt.plot(range(1, len(diff)+1), diff, label=f"Step size = {step_size}")

plt.xlabel("$k$")
plt.ylabel("$f(x_{k-1}) - f(x_{k})$")
plt.title("$f(x_{k-1}) - f(x_{k})$ v.s. $k$ for Difference Step Sizes")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Surface plot methods from matplotlib

# Generate synthetic data
np.random.seed(42) # Last two digits of ID
X = 2 * np.random.rand(100, 2) # Generate 100 random 2D data points
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100) # Generate y values with noise

# Add a column of ones to X for the bias terms (intercept) in the normal equation
x = np.c_[np.ones((100, 2)), X]

# Find theta using the normal equation
theta = np.linalg.pinv(x.T @ x) @ x.T @ y

# Print the intercept and coefficients of the regression plane
print("Intercept:", theta[0])
print("x1 coefficient:", theta[1])
print("x2 coefficient:", theta[2])

# Plot the generated data and the regression plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of data
ax.scatter(X[:, 0], X[:, 1], y, c='blue', marker='o', alpha=0.6)

# Create a meshgrid for the plane
x1_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 20)
x2_vals = np.linspace(X[:, 1].min(), X[:, 1].max(), 20)
x1_mesh, x2_mesh = np.meshgrid(x1_vals, x2_vals)

# Regression plane: z = theta0 + theta1*x1 + theta2*x2
z_mesh = theta[0] + theta[1]*x1_mesh + theta[2]*x2_mesh

# Plot the plane
ax.plot_surface(x1_mesh, x2_mesh, z_mesh, alpha=0.4, color='red')

# Labels
ax.set_xlabel("X₁")
ax.set_ylabel("X₂")
ax.set_zlabel("y")
ax.set_title("Linear Regression Plane Fit")

plt.show()
```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. Your task is to

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logistic function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate training and testing datasets
np.random.seed(595)

# Generate 2D feature matrix with 100 points
X = np.random.rand(2, 100)

# True underlying parameters used to synthesize labels
w_true = np.array([1.5, -2.5])
b_true = 1.0  # True bias
probabilities = sigmoid(np.dot(w_true.T, X) + b_true) # Compute probability using true model
Y = (probabilities > 0.5).astype(int) # Convert probabilities into binary labels 
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100) # Add noise

# Split training and test sets 
X_train, X_test = X[:,:80], X[:,80:]
Y_train, Y_test = Y[:80], Y[80:]

# Set hyperparameters and initialize parameters
w = np.zeros(X.shape[0])
b = 0.0

# Tuned hyperparameters
learning_rate = 0.5 # Prevent overshooting
num_epochs = 500 # Lower epochs to prevent overfitting

# Start training
for epoch in range(num_epochs):
    A_train = sigmoid(np.dot(w.T, X_train) + b) # Forward pass — compute predictions for training data

    # Compute gradients of the cost function
    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)
    dJdb = np.mean(A_train - Y_train)

    # Update parameters using gradient descent
    w -= learning_rate * dJdw
    b -= learning_rate * dJdb

# Predict on training set
A_train = sigmoid(np.dot(w.T, X_train) + b)
predictions_train = (A_train > 0.5).astype(int)

# Predict on test set
A_test = sigmoid(np.dot(w.T, X_test) + b)
predictions_test = (A_test > 0.5).astype(int)

# Calculate training and testing accuracy
train_accuracy = np.mean(predictions_train == Y_train)
test_accuracy = np.mean(predictions_test == Y_test)

print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")

# Plot the training dataset and regression decision boundary
plt.scatter(X[0, :80], X[1,:80], c=Y[:80], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
ax = plt.gca()

xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create a grid over the feature space
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))

# Compute decision boundary values
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z)
Z = Z.reshape(xx.shape)

# Plot contour where sigmoid = 0.5 (decision boundary)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```

```python
import h5py
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import matplotlib.pyplot as plt
```

```python
# Load h5 training datset
def load_h5_dataset(path, x_dataset_name, y_dataset_name):
    with h5py.File(path, "r") as f:
        X = np.array(f[x_dataset_name][:])
        Y = np.array(f[y_dataset_name][:]).reshape(-1)
    return X, Y

# Convert np arrays to torch DataLoader
def make_loader(X, Y, batch_size=32, shuffle=True):
    X_t = torch.tensor(X, dtype=torch.float32) / 255.0 # Convert to torch tensor and normalize images
    X_t = X_t.permute(0, 3, 1, 2) # Convert to NCHW
    Y_t = torch.tensor(Y, dtype=torch.long) # Make labels as long integers
    dataset = TensorDataset(X_t, Y_t) # Create a dataset and loader
    return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)

# Define CNN
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            # First convolution layer: 3 input channels to 16 output channels
            nn.Conv2d(3, 16, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2), # Downsample by factor of 2

            # Second convolution layer: 16 to 32 channels
            nn.Conv2d(16, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2), # Downsample by factor of 2

            # Flatten feature maps into a vector
            nn.Flatten(),

            # Fully-connected layer
            nn.Linear(32 * 16 * 16, 64),
            nn.ReLU(),

            # Final output layer (binary classification → 2 outputs)
            nn.Linear(64, 2)
        )

    def forward(self, x):
        return self.net(x)
    
# Train model
def train_model(model, train_loader, lr=0.001, epochs=10):
    criterion = nn.CrossEntropyLoss() # Loss for classification
    optimizer = optim.Adam(model.parameters(), lr=lr)

    for epoch in range(epochs): # Set model to training mode
        model.train()
        for Xb, yb in train_loader:
            optimizer.zero_grad() # Reset gradients
            loss = criterion(model(Xb), yb) # Compute loss and forward pass
            loss.backward() # Backprop
            optimizer.step() # Update weights

# Compute accuracy and get misclassified indices
def evaluate_model(model, loader):
    model.eval() # Inference mode
    all_preds = []
    all_labels = []

    with torch.no_grad(): # Disable gradient tracking
        for Xb, yb in loader:
            preds = torch.argmax(model(Xb), dim=1)
            all_preds.append(preds)
            all_labels.append(yb)

    preds = torch.cat(all_preds)
    labels = torch.cat(all_labels)
    
    accuracy = (preds == labels).float().mean().item() # Compute accuracy
    misclassified = (preds != labels).nonzero().flatten().numpy() # Get misclassified sample indices

    return accuracy, misclassified, preds, labels

# Display sample images
def show_images(X, indices, title):
    plt.figure(figsize=(8, 4))
    for i, idx in enumerate(indices[:4]):
        plt.subplot(1, 4, i+1)
        plt.imshow(X[idx])
        plt.title(f"Idx {idx}")
        plt.axis("off")
    plt.suptitle(title)
    plt.show()
```

```python
X_train, Y_train = load_h5_dataset("train_catvnoncat.h5",'train_set_x','train_set_y')
X_test, Y_test = load_h5_dataset("test_catvnoncat.h5",'test_set_x','test_set_y')

# Convert to DataLoaders
train_loader = make_loader(X_train, Y_train)
test_loader = make_loader(X_test, Y_test, shuffle=False)


# Initialize and train model
model = CNN()
train_model(model, train_loader, epochs=10)

# Evaluate original model
train_acc, _, _, _ = evaluate_model(model, train_loader)
test_acc, mis_idx, _, _ = evaluate_model(model, test_loader)

print("Training accuracy:", train_acc)
print("Test accuracy:", test_acc)
print("Misclassified indices:", mis_idx)

# Show 4 failed cases
if len(mis_idx) > 0:
    show_images(X_test, mis_idx, "Original Test Set: Misclassified Samples")
```

```python
# Re-split the training set into new train/test sets
X_new_train = X_train[:162]
Y_new_train = Y_train[:162]

X_new_test = X_train[162:] # 209 total to 47 test images
Y_new_test = Y_train[162:]

# Create new DataLoaders
new_train_loader = make_loader(X_new_train, Y_new_train)
new_test_loader = make_loader(X_new_test, Y_new_test, shuffle=False)

# Reinitialize model
model2 = CNN()
train_model(model2, new_train_loader, epochs=10)

# Evaluate modified dataset
new_train_acc, _, _, _ = evaluate_model(model2, new_train_loader)
new_test_acc, new_mis_idx, _, _ = evaluate_model(model2, new_test_loader)

print("New training accuracy:", new_train_acc)
print("New test accuracy:", new_test_acc)
print("Misclassified indices (modified test set):", new_mis_idx)

# Show 4 failed cases
if len(new_mis_idx) > 0:
    show_images(X_new_test, new_mis_idx, "Modified Test Set: Misclassified Samples")
```
